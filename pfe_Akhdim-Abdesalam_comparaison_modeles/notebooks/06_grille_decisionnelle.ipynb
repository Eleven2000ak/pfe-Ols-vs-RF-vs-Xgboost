#  Grille décisionnelle et validation finale des hypothèses

---

## 1. Objectif de la grille décisionnelle

La grille décisionnelle permet de recommander un modèle de régression selon plusieurs critères :

- linéarité des relations
- taille de l’échantillon $n$
- besoin d’interprétabilité
- tolérance au risque

Elle repose sur les résultats de trois hypothèses testées sur données synthétiques et réelles.

---

## 2. Synthèse des hypothèses

---

### H1 – Linéarité du processus génératif

**Objectif** : vérifier si la relation entre variables est linéaire.

**Méthode** : comparaison des performances entre :

- régression linéaire (OLS)
- Random Forest

Si les performances sont proches → linéarité  
Sinon → non-linéarité

**Résultat** :

$H1 = \text{False}$

→ le processus est **non linéaire**

---

### H2 – Nécessité de modèles complexes

**Objectif** : tester si les modèles non linéaires améliorent les performances sur données réelles.

**Méthode** : comparaison OLS vs Random Forest / XGBoost via :

- $R^2$
- MAE

**Résultat** :

$H2 = \text{False}$

→ les modèles complexes sont nécessaires

---

### H3 – Robustesse aux outliers

**Objectif** : mesurer la sensibilité des modèles aux valeurs aberrantes.

**Méthode** :

- ajout de $5\%$ d’outliers
- comparaison de la dégradation des performances

**Résultat** :

$H3 = \text{True}$

→ Random Forest et XGBoost sont robustes

---
